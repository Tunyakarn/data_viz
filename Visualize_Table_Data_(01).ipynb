```
import pandas as pd
```

```
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df
```

### à¹à¸›à¸¥à¸‡ data à¸ˆà¸²à¸ pandas dataframe à¹€à¸›à¹‡à¸™ numpy array

```
np_data = df.iloc[:,:-1].to_numpy()
np_data
```

### Vizsualize array data à¸”à¹‰à¸§à¸¢ pixel

```
matplotlib.rcParams['figure.figsize']=[15,15] #à¸à¸³à¸«à¸™à¸”à¸‚à¸™à¸²à¸”à¸‚à¸­à¸‡ figure [à¸à¸§à¹‰à¸²à¸‡,à¸¢à¸²à¸§]
plt.imshow(np_data[:50,:],interpolation='nearest')
```
