```
import pandas as pd
```

```
df = pd.read_csv('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/io/data/csv/iris.csv')
df
```

## Parallel Coordinates

```
pd.plotting.parallel_coordinates(df, 'Name');
```

### ปรับ parameter: color

```
pd.plotting.parallel_coordinates(df, 'Name', color=('r','g','b')) #ตัวย่อของสี R G B W C M Y K
```

```
pd.plotting.parallel_coordinates(df, 'Name', color=((1,0.8,0),(0.2,0.4,0.5),(0.6,0,1))) #rgb values (r,g,b) | r,g,b -> [0,1] #สามารถกำหนดปริมาณของแสงที่มาผสมกันกัน
```

### ปรับ parameter: colormap

```
pd.plotting.parallel_coordinates(df, 'Name', colormap='cool')
```

```
pd.plotting.parallel_coordinates(df, 'Name', colormap='tab10')
```

## เริ่ม Matplotlip

```
from matplotlib import pyplot as plt #from เป็นโฟลอเดอร์ใหญ่เก็บรวมแพคเกจหลายๆแพคเกจไว้ เราจะ import มาเฉพาะ pyplot แล้วเปลี่ยนชื่อ pyplot เป็น plt
```

### แปลง data จาก pandas dataframe เป็น numpy array

```
np_data = df.iloc[:,:-1].to_numpy()
np_data
```

### Vizsualize array data ด้วย pixel

```
import matplotlib
```

```
matplotlib.rcParams['figure.figsize']=[15,15] #กำหนดขนาดของ figure [กว้าง,ยาว]
plt.imshow(np_data[:50,:],interpolation='nearest')
```

```
plt.imshow(np_data[50:100,:],interpolation='nearest')
```

### การใช้ subplot

```
plt.subplot(1,2,1) #subplot(มีกี่แถว กี่หลัก เป็นตัวที่เท่าไหร่)
plt.imshow(np_data[:50,:],interpolation='nearest', cmap='magma')
plt.subplot(1,2,2)
plt.imshow(np_data[50:100,:],interpolation='nearest', cmap='magma')
```

```
plt.imshow(np_data[:10,:],interpolation='nearest', cmap='magma', alpha= 0.5) #alpha คือ ความโปร่งใส
```
